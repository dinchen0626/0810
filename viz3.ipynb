```python
df = df.set_index('Date')
```

```python
df['Date']=df.index
```

```python
import plotly.graph_objects as go
line1 = go.Scatter(x=df['Date'], y=df['Close'], name='Close')
fig  = go.Figure(line1)
fig.show()
```

```python
import plotly.graph_objects as go
line1 = go.Scatter(x=df['Date'], y=df['Close'], name='Close')
line2 = go.Scatter(x=df['Date'], y=df['Low'], name='Low')
fig  = go.Figure([line1, line2])
fig.update_layout(
    title = "Amazon Stock Price",
    xaxis_title = "Date",
    yaxis_title = "Price"
)
fig.show()
```
